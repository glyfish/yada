## Research Notes Metadata

```python
collection = vs._collection
probe = collection.get(limit=100000)
metas  = [m for m in probe.get("metadatas", []) if m]
len(metas)
```

```python
titles = Counter(m.get("title") for m in metas if "title" in m and m.get("title"))
paths = Counter(m.get("path") for m in metas if "path" in m and m.get("path"))

for title in titles:
    print(f"title {title}")

for path in paths:
    print(f"path: {path}")
```
