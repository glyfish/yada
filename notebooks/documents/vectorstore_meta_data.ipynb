```python
%reload_ext autoreload
%autoreload 2

import sys
import os
from collections import Counter

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '../..')))

from apps.agentic.core.document_loaders.research_library_document_loader import ResearchLibraryChromaDocumentLoader
from apps.agentic.core.document_loaders.document_library_loader import DocumentLibraryLoader
from apps.agentic.agents.document.document_store_info_agent import DocumentStoreInfoAgent
from apps.agentic.core.utils import (set_chatgpt_env, set_langsmith_env)
from apps.agentic.core.constants import (RESEARCH_LIBRARY_DB_NAME, RESEARCH_LIBRARY_COLLECTION_NAME, 
                                         PDF_DOCUMENT_LIBRARY_DB_NAME, PDF_DOCUMENT_LIBRARY_COLLECTION_NAME)

DB_PATH = "../../.db"

set_chatgpt_env(filedir="../../.keys")
set_langsmith_env(filedir="../../.keys")
```

## Research Notes 

```python
RESEARCH_LIBRARY_DB_NAME, RESEARCH_LIBRARY_COLLECTION_NAME, DB_PATH, os.listdir(DB_PATH)
```

```python
vs = ResearchLibraryChromaDocumentLoader(DB_PATH).vectorstore
collection = vs._collection

probe = collection.get(limit=5000)  # adjust as needed
metas  = [m for m in probe.get("metadatas", []) if m]
len(metas)
```

```python
titles = Counter(m.get("title") for m in metas if "title" in m and m.get("title")).most_common(10)
paths = Counter(m.get("path") for m in metas if "path" in m and m.get("path")).most_common(10)
n = min(len(titles), len(paths))

for i in range(n):
    print(f"title counts: {titles[i]}, path counts: {paths[i]}")
```
