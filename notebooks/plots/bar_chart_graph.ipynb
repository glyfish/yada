## Top 10 Countries by GDP

The GDP for the 10 largest country economies in US dollars.

|Country|GDP (T US$)|
|-------|----------|
|US|30.51|
|China|19.23|
|Germany|4.74|
|India|4.19|
|Japan|4.18|
|UK|3.84|
|France|3.21|
|Italy|2.42|
|Canada|2.23|

```python
data = {
    'USA': 30.51, 
    'China': 19.23, 
    'Germany': 4.74,
    'India': 4.19,
    'Japan': 4.18,
    'UK': 3.84,
    'France': 3.21,
    'Italy': 2.42,
    'Canada': 2.23,
    'Brazil': 2.12,
}

x = numpy.array(list(data.keys()))
y = data.values()

bar(y, x, alpha=1.0, bar_width=0.9, xlabel_rotation=45,
    xlabel='Countries', ylabel='GDP (in trillion USD)',title='Top 10 Countries by GDP in 2025',
    figsize=(10, 6))
```
