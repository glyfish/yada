## Bar Plot

```python
data = {
    'USA': 30.51, 
    'China': 19.23, 
    'Germany': 4.74,
    'India': 4.19,
    'Japan': 4.18,
    'UK': 3.84,
    'France': 3.21,
    'Italy': 2.42,
    'Canada': 2.23,
    'Brazil': 2.12,
}

x = numpy.array(list(data.keys()))
y = data.values()

bar(y, x, alpha=1.0, bar_width=0.9, xlabel_rotation=45,
    xlabel='Countries', ylabel='GDP (in trillion USD)',title='Top 10 Countries by GDP in 2025',
    figsize=(10, 6))
```

## Multibar Example

```python
year_data = {'2025':
                {
                    'USA': 30.51, 
                    'China': 19.23, 
                    'Germany': 4.74,
                    'India': 4.19,
                    'Japan': 4.18,
                    'UK': 3.84,
                    'France': 3.21,
                    'Italy': 2.42,
                    'Canada': 2.23,
                    'Brazil': 2.12,
                },
        '2024': 
                {
                    'USA': 24.51, 
                    'China': 17.23, 
                    'Germany': 3.74,
                    'India': 3.19,
                    'Japan': 3.18,
                    'UK': 2.84,
                    'France': 2.21,
                    'Italy': 2.12,
                    'Canada': 1.93,
                    'Brazil': 1.8,
                },
        '2023': 
            {
                'USA': 36.51, 
                'China': 18.23, 
                'Germany': 4.54,
                'India': 4.35,
                'Japan': 4.21,
                'UK': 3.74,
                'France': 3.01,
                'Italy': 2.02,
                'Canada': 1.83,
                'Brazil': 1.12,
        },
}

labels = list(year_data.keys())
values = [numpy.array(list(year_data.values())) for year_data in year_data.values()]
x = numpy.array(list(list(year_data.values())[0].keys()))
y = values

multibar(y, x, alpha=1.0, bar_width=0.9, xlabel_rotation=45, labels=labels,
         xlabel='Countries', ylabel='GDP (in trillion USD)',title='Top 10 Countries by GDP in 2025',
         figsize=(10, 6))
```
