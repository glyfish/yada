## Single Time Series

```python
time = [datetime.datetime(1960, 1, 1, 0, 0), datetime.datetime(1970, 1, 1, 0, 0),
        datetime.datetime(1980, 1, 1, 0, 0), datetime.datetime(1990, 1, 1, 0, 0),
        datetime.datetime(2000, 1, 1, 0, 0), datetime.datetime(2010, 1, 1, 0, 0),
        datetime.datetime(2020, 1, 1, 0, 0), datetime.datetime(2021, 1, 1, 0, 0),
        datetime.datetime(2022, 1, 1, 0, 0), datetime.datetime(2023, 1, 1, 0, 0)]

values = [0.543, 1.076, 2.857, 5.963, 10.285, 15.049, 21.323, 23.594, 25.744, 27.721]
```

```python
curve(numpy.array(values), numpy.array(time), xlabel="Year", ylabel="Value", title="Time Series Example", figsize=(10, 6))
```

## Stack Plot

```python
time = numpy.array([datetime.datetime(i, 1, 1, 0, 0) for i in range(1960, 2024)])
values = [numpy.random.uniform(0, 30, len(time)), 
          numpy.random.uniform(0, 30, len(time)), 
          numpy.random.uniform(0, 30, len(time))]
```

```python
stack(values, time, xlabel="Year", ylabels=["Value-1", "Value-2", "Value-3"], title="Time Series Example", figsize=(10, 6),
      labels=["Series 1", "Series 2", "Series 3"])
```

## Comparison Plot

```python
time = [datetime.datetime(1960, 1, 1, 0, 0), datetime.datetime(1970, 1, 1, 0, 0),
        datetime.datetime(1980, 1, 1, 0, 0), datetime.datetime(1990, 1, 1, 0, 0),
        datetime.datetime(2000, 1, 1, 0, 0), datetime.datetime(2010, 1, 1, 0, 0),
        datetime.datetime(2020, 1, 1, 0, 0), datetime.datetime(2021, 1, 1, 0, 0),
        datetime.datetime(2022, 1, 1, 0, 0), datetime.datetime(2023, 1, 1, 0, 0)]

vals = numpy.array([0.543, 1.076, 2.857, 5.963, 10.285, 15.049, 21.323, 23.594, 25.744, 27.721])
values = [vals, vals[::-1]]
```

```python
comparison(values, numpy.array(time), xlabel="Year", ylabel="Value", title="Multiple Time Series Example", 
          labels=["Data 1", "Data 2"], figsize=(10, 6))
```
